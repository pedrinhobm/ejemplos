1. implementa una funcion en python que devuelva el valor de la serie y los valores valores parciales de la serie. La funcion debe estar basada en un bucle for. Solo debe tener como parametro la cantidad de elementos. Presente su resultado para 4000 elementos , y muestre una grafica de como cambiaron los valores parciales de la serie con el numero de elementos

```python
import matplotlib.pyplot as plt

def sumatoria_for(num):
    rpta = 0
    valor = []
    for i in range(1, num + 1):
        term = 1 / i**2
        rpta += term
        valor.append(rpta)
    return rpta, valor
```

```python
# Calcula la serie con 4000 elementos
num = 4000
suma, valor = sumatoria_for(num)

print("El valor de la serie de ",num, "es" ,suma)
# Imprime el valor de la serie
print("Los valores de la serie son")
print(valor)
```

```python
# Grafica los valores parciales de la serie
plt.plot(range(1, num + 1), valor)
plt.xlabel("Nmero de elementos")
plt.ylabel("Valores parciales de la serie")
plt.title("Valores parciales de la serie de pi^2/6")
plt.grid(True)
plt.show()
```

2. Implementa una funcion en python que devuelva el valor de la serie y los valores valores parciales de la serie. La funcion debe estar basada en un bucle while. Solo debe tener como parametros  un valor de referencia , la precision y la cantidad tope de iteraciones. Puede usar el valor de Pi proporcionado por el modulo math para tener una referencia. Presente su resultado para una presion de 1e-4

```python
import math

def sumatoria_while(referencia, precision, max_iter):
    suma = 0
    valores = []
    i = 1

    while True:
        suma += 1 / (i**2)
        valores.append(suma)

        if len(valores) > 2:
            error = abs(valores[-1] - referencia) / referencia
            if error < precision:
                break
        if i >= max_iter:
            break
        i +=1

    return suma, valores
```

```python
referencia = math.pi**2 / 6
precision = 1e-4
max_iter = 4000

# Calcula la serie con los parmetros especificados
rpta, valores = sumatoria_while(referencia, precision, max_iter)
```

```python
# Imprime el valor de la serie y la cantidad de iteraciones realizadas
print("Valor de la serie:", rpta)
print("Cantidad de iteraciones:", len(valores))
print(valores)
```

3. implementa una funcion en python que devuelva el valor de la serie y los valores valores parciales de la serie. La funcion debe estar basada en un bucle while. Solo debe tener como parametro la precision y la cantidad tope de iteraciones. Presente su resultado para una presion de 1e-5

```python
import math

def sumatoria_while2(precision, max_iter):
    sum = 0
    values = []
    i = 1

    while True:
        sum += 1 / (i**2)
        values.append(sum)

        if len(values) > 2:
            error = abs(values[-1] - values[-2]) / values[-2]
            if error < precision:
                break
        if i >= max_iter:
            break
        i +=1

    return sum, values
```

```python
precision = 1e-5
max_iter = 4000

sum, values = sumatoria_while2(precision, max_iter)

# Imprime el valor de la serie y la cantidad de iteraciones realizadas
print("Valor de la serie:", sum)
print("Cantidad de iteraciones:", len(values))
print(values)
```
